```python
# استيراد المكتبات التي سنستخدمها

import sqlite3   # للتواصل مع قاعدة البيانات SQLite
from faker import Faker   # لتوليد بيانات عشوائية مثل الأسماء والتواريخ
import random   # لتوليد أرقام وقيم عشوائية

# إنشاء مولد بيانات من مكتبة Faker
fake = Faker()

print("Libraries loaded successfully")
```

```python
# الاتصال بقاعدة البيانات الموجودة في مجلد database

connection = sqlite3.connect('../database/fleetpulse.db')

# إنشاء cursor لتنفيذ أوامر SQL
cursor = connection.cursor()

print("Connected to database successfully")
```

```python
# توليد بيانات السائقين

drivers_data = []

for i in range(1, 101):  # إنشاء 100 سائق

    full_name = fake.name()  # اسم عشوائي
    national_id = str(random.randint(1000000000, 9999999999))  # رقم هوية عشوائي
    phone_number = "05" + str(random.randint(10000000, 99999999))  # رقم هاتف سعودي
    license_number = "LIC" + str(random.randint(10000, 99999))  # رقم رخصة
    license_type = random.choice(["Light", "Medium", "Heavy"])  # نوع الرخصة
    hire_date = fake.date()  # تاريخ التوظيف
    city = fake.city()  # مدينة عشوائية
    rating = round(random.uniform(3.0, 5.0), 1)  # تقييم السائق
    status = "Active"  # حالة السائق

    drivers_data.append((
        i,
        full_name,
        national_id,
        phone_number,
        license_number,
        license_type,
        hire_date,
        city,
        rating,
        status
    ))

cursor.executemany("""
INSERT INTO Drivers (
driver_id,
full_name,
national_id,
phone_number,
license_number,
license_type,
hire_date,
city,
rating,
status
)
VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
""", drivers_data)

connection.commit()

print("Drivers inserted successfully")
```

```python
# توليد بيانات المركبات

vehicle_brands = ["Volvo", "Mercedes", "Scania", "MAN", "DAF", "Iveco"]
vehicle_types = ["Truck", "Van", "Trailer"]

vehicles_data = []

for i in range(2, 51):  # إنشاء 50 مركبة

    plate_number = f"TRK{i:04d}"  # رقم لوحة المركبة
    vehicle_type = random.choice(vehicle_types)  # نوع المركبة
    brand = random.choice(vehicle_brands)  # الشركة المصنعة
    model = f"Model-{random.randint(100,999)}"  # موديل المركبة
    manufacture_year = random.randint(2015, 2024)  # سنة الصنع
    capacity_tons = random.randint(5, 20)  # سعة الحمولة بالطن
    fuel_type = "Diesel"  # نوع الوقود
    purchase_date = fake.date()  # تاريخ الشراء
    status = "Available"  # حالة المركبة

    vehicles_data.append((
        i,
        plate_number,
        vehicle_type,
        brand,
        model,
        manufacture_year,
        capacity_tons,
        fuel_type,
        purchase_date,
        status
    ))

cursor.executemany("""
INSERT INTO Vehicles (
vehicle_id,
plate_number,
vehicle_type,
brand,
model,
manufacture_year,
capacity_tons,
fuel_type,
purchase_date,
status
)
VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
""", vehicles_data)

connection.commit()

print("Vehicles inserted successfully")
```

```python
# توليد بيانات الرحلات
# نبدأ من trip_id = 2 لأن الرحلة رقم 1 موجودة مسبقًا من الإدخال اليدوي

trips_data = []

for i in range(2, 1002):  # 1000 رحلة

    driver_id = random.randint(1, 100)   # اختيار سائق عشوائي من السائقين الموجودين
    vehicle_id = random.randint(1, 50)   # اختيار مركبة عشوائية من المركبات الموجودة
    route_id = random.randint(1, 26)     # اختيار مسار عشوائي من المسارات الموجودة

    departure_datetime = fake.date_time_between(start_date='-1y', end_date='now')
    
    # مدة الرحلة الفعلية بالساعات (تقريبية)
    trip_duration_hours = random.randint(1, 12)
    
    # وقت الوصول = وقت الانطلاق + مدة الرحلة
    arrival_datetime = departure_datetime + __import__("datetime").timedelta(hours=trip_duration_hours)

    trip_status = random.choice(["Completed", "Delayed", "Cancelled"])
    shipment_count = random.randint(1, 5)
    total_weight_kg = random.randint(500, 20000)
    delivered_on_time = random.choice([0, 1])  # 1 = نعم، 0 = لا

    trips_data.append((
        i,
        driver_id,
        vehicle_id,
        route_id,
        departure_datetime.strftime("%Y-%m-%d %H:%M:%S"),
        arrival_datetime.strftime("%Y-%m-%d %H:%M:%S"),
        trip_status,
        shipment_count,
        total_weight_kg,
        delivered_on_time
    ))

cursor.executemany("""
INSERT INTO Trips (
trip_id,
driver_id,
vehicle_id,
route_id,
departure_datetime,
arrival_datetime,
trip_status,
shipment_count,
total_weight_kg,
delivered_on_time
)
VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
""", trips_data)

connection.commit()

print("Trips inserted successfully")
```

```python
# توليد بيانات الشحنات
# نبدأ من shipment_id = 2 لأن الشحنة رقم 1 موجودة مسبقًا

shipments_data = []
shipment_id_counter = 2

for trip_id in range(1, 1002):  # من الرحلة 1 إلى 1001 تقريبًا
    num_shipments = random.randint(1, 4)  # كل رحلة فيها من 1 إلى 4 شحنات

    for _ in range(num_shipments):
        customer_name = fake.company()
        pickup_city = fake.city()
        delivery_city = fake.city()
        weight_kg = random.randint(100, 5000)
        shipment_value = random.randint(5000, 50000)
        delivery_status = random.choice(["Delivered", "Delayed", "Cancelled"])
        delivery_deadline = fake.date()

        shipments_data.append((
            shipment_id_counter,
            trip_id,
            customer_name,
            pickup_city,
            delivery_city,
            weight_kg,
            shipment_value,
            delivery_status,
            delivery_deadline
        ))

        shipment_id_counter += 1

cursor.executemany("""
INSERT INTO Shipments (
shipment_id,
trip_id,
customer_name,
pickup_city,
delivery_city,
weight_kg,
shipment_value,
delivery_status,
delivery_deadline
)
VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
""", shipments_data)

connection.commit()

print("Shipments inserted successfully")
```

```python
# توليد بيانات سجلات الوقود
# نبدأ من fuel_log_id = 2 لأن السجل رقم 1 موجود مسبقًا

fuel_logs_data = []
fuel_log_id_counter = 2

for trip_id in range(1, 1002):  # المرور على كل الرحلات
    num_fuel_logs = random.randint(1, 2)  # كل رحلة لها 1 أو 2 سجل وقود

    for _ in range(num_fuel_logs):
        vehicle_id = random.randint(1, 50)   # اختيار مركبة عشوائية
        fuel_date = fake.date()
        liters_filled = round(random.uniform(50, 300), 2)   # كمية الوقود باللتر
        fuel_cost = round(liters_filled * random.uniform(1.8, 2.8), 2)  # تكلفة تقريبية
        odometer_km = random.randint(10000, 300000)   # قراءة العداد
        station_city = fake.city()

        fuel_logs_data.append((
            fuel_log_id_counter,
            trip_id,
            vehicle_id,
            fuel_date,
            liters_filled,
            fuel_cost,
            odometer_km,
            station_city
        ))

        fuel_log_id_counter += 1

cursor.executemany("""
INSERT INTO Fuel_Logs (
fuel_log_id,
trip_id,
vehicle_id,
fuel_date,
liters_filled,
fuel_cost,
odometer_km,
station_city
)
VALUES (?, ?, ?, ?, ?, ?, ?, ?)
""", fuel_logs_data)

connection.commit()

print("Fuel logs inserted successfully")
```

```python
# توليد بيانات الصيانة
# نبدأ من maintenance_id = 2 لأن السجل رقم 1 موجود مسبقًا

maintenance_data = []
maintenance_id_counter = 2

maintenance_types = [
    "Oil Change",
    "Tire Replacement",
    "Brake Repair",
    "Engine Check",
    "Battery Replacement",
    "Transmission Repair"
]

service_providers = [
    "Volvo Service Center",
    "Mercedes Workshop",
    "Scania Maintenance Hub",
    "National Auto Service",
    "TruckCare Center"
]

for i in range(2, 202):  # 200 سجل صيانة تقريبًا

    vehicle_id = random.randint(1, 50)  # اختيار مركبة عشوائية
    maintenance_date = fake.date()
    maintenance_type = random.choice(maintenance_types)
    cost = round(random.uniform(500, 8000), 2)
    service_provider = random.choice(service_providers)
    downtime_days = random.randint(1, 7)

    maintenance_data.append((
        maintenance_id_counter,
        vehicle_id,
        maintenance_date,
        maintenance_type,
        cost,
        service_provider,
        downtime_days
    ))

    maintenance_id_counter += 1

cursor.executemany("""
INSERT INTO Maintenance (
maintenance_id,
vehicle_id,
maintenance_date,
maintenance_type,
cost,
service_provider,
downtime_days
)
VALUES (?, ?, ?, ?, ?, ?, ?)
""", maintenance_data)

connection.commit()

print("Maintenance records inserted successfully")
```

```python
# توليد بيانات المخالفات
# نبدأ من violation_id = 2 لأن السجل رقم 1 موجود مسبقًا

violations_data = []
violation_id_counter = 2

violation_types = [
    "Speeding",
    "Overloading",
    "Signal Violation",
    "Unsafe Driving",
    "Parking Violation",
    "Documentation Issue"
]

for i in range(2, 302):  # حوالي 300 مخالفة

    trip_id = random.randint(1, 1001)
    driver_id = random.randint(1, 100)
    vehicle_id = random.randint(1, 50)

    violation_date = fake.date()
    violation_type = random.choice(violation_types)
    fine_amount = round(random.uniform(100, 2000), 2)

    violations_data.append((
        violation_id_counter,
        trip_id,
        driver_id,
        vehicle_id,
        violation_date,
        violation_type,
        fine_amount
    ))

    violation_id_counter += 1

cursor.executemany("""
INSERT INTO Violations (
violation_id,
trip_id,
driver_id,
vehicle_id,
violation_date,
violation_type,
fine_amount
)
VALUES (?, ?, ?, ?, ?, ?, ?)
""", violations_data)

connection.commit()

print("Violations inserted successfully")
```
